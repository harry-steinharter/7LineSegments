```python
import re
import pandas as pd
import numpy as np
import os
```

```python
trials = {'Sub_Number':[], 'TrialStart':[], 'TrialEnd':[], 'Trial_Number':[], 'Condition':[], 'Target_Contrast':[]}
saccades = {'Sub_Number':[], 'StartTime': [], 'EndTime': [], 'Duration':[], 'StartX':[], 'StartY':[], 'EndX':[], 'EndY':[], 'Distance':[], 'PeakVelocity':[]}
files = [file for file in os.listdir() if file.endswith('.asc') and file != '75.asc']
file = files[3]
with open(file,'r') as f:
    lines = f.readlines()
```

```python
for line in lines:
    if "BUTTON" in line:
        print(line)
```

```python
for file in files:
    with open(file,'r') as f:
        lines = f.readlines()

    subNumber = file.split('.')[0]
    
    for line in lines:
        if "TRIAL_START"  in line:
            match = re.search(r'MSG\s+(\d+)\s+TRIAL_START\s+(\d+)\s*\|\s*([\w]+)\s*\|\s*([\d.]+)', line)
            T, N, C, TC = match.groups()
            trials['Sub_Number'].append(int(subNumber))
            trials['TrialStart'].append(int(T))
            trials['Trial_Number'].append(int(N))
            trials['Condition'].append(C)
            trials['Target_Contrast'].append(float(TC))
        elif "TRIAL_END" in line:
            trials['TrialEnd'].append(line.split(' ')[0].split()[1])
        elif "ESACC" in line:
            pattern = r"ESACC\s+[LR]\s+([\d.eE+-]+|\.)\s+([\d.eE+-]+|\.)\s+([\d.eE+-]+|\.)\s+([\d.eE+-]+|\.)\s+([\d.eE+-]+|\.)\s+([\d.eE+-]+|\.)\s+([\d.eE+-]+|\.)\s+([\d.eE+-]+|\.)\s+([\d.eE+-]+|\.)"
            match = re.search(pattern, line)

            values = np.array(match.groups())
            NaNs = [i for i, val in enumerate(values) if val == '.']
            Es = [i for i, val in enumerate(values) if val.__contains__('e')]
            values[NaNs] = -99.
            values[Es] = -99.
            values = np.asarray(values,dtype=float)

            ST, ET, Durt, SX, SY, EX, EY, Dist, V = values
            saccades['Sub_Number'].append(int(subNumber))
            saccades['StartTime'].append(ST)
            saccades['EndTime'].append(ET)
            saccades['Duration'].append(Durt)
            saccades['StartX'].append(SX)
            saccades['StartY'].append(SY)
            saccades['EndX'].append(EX)
            saccades['EndY'].append(EY)
            saccades['Distance'].append(Dist)
            saccades['PeakVelocity'].append(V)
```

```python
for key in trials.keys():
    print(key,len(trials[key]))
```

```python
saccades_df = pd.DataFrame(saccades)
trials_df = pd.DataFrame(trials)
```

```python

saccades_df.to_csv("saccades_df.csv",index=False)
trials_df.to_csv("trials_df.csv",index=False)
```
